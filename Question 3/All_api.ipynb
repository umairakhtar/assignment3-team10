```
import pandas as pd
google = pd.read_csv('google_result.csv')
ibm = pd.read_csv('ibm_result.csv')
azure = pd.read_csv('azure_result.csv')
```

```
print(google.shape)
print(ibm.shape)
print(azure.shape)
```

```
google_score = list(google.score)
google_label = list(google.predict)
original_label = list(google.label)
google.head()
```

###Confusion Matrix from Google API

```
from sklearn.metrics import accuracy_score
print('accuracy is: ',accuracy_score(google.label,google.predict))
pd.crosstab(google.label,google.predict)
```

```
ibm['score'] = ibm['score'].replace('neutral',0)
ibm_score = list(ibm.score)
ibm_label = list(ibm.predict)
ibm.head()
```

###Confusion Matrix for Watson API

```
print('accuracy is: ',accuracy_score(google.label,ibm.predict) )
pd.crosstab(google.label,ibm.predict)
```

```
azure_score = list(azure.score)
azure_label = list(azure.sentiment_predicted)
azure.head()
```

###Confusion Matrix of Azure

```
print('accuracy is: ',accuracy_score(google.label,azure.sentiment_predicted) )
pd.crosstab(google.label,azure.sentiment_predicted)
```

```
dict_labels = {'google': google_label,'ibm': ibm_label, 'azure': azure_label }
df_labels = pd.DataFrame(dict_labels)
df_labels.head()
```

```
df_labels = df_labels.replace({'positive':0,'negative':1,'neutral':2})
df_labels.head()
```

```
mode = pd.DataFrame(df_labels.mode(axis=1),)
mode.columns=['mode1','mode2','mode3']
```

###Average Accuracy of All API

```
print('accuracy of all API average is :', accuracy_score(mode['mode1'],google['label'].replace({'positive':0,'negative':1,'neutral':2})))
pd.crosstab(mode['mode1'],google['label'].replace({'positive':0,'negative':1,'neutral':2}))
```
